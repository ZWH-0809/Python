```python
import time
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

t = np.linspace(0, 10, 1000)
N0=100
r=1.2
K=10000
N=K/(1+(K-N0)/N0*np.exp(-r*t))
plt.figure(figsize=(6,4))
plt.plot(t,N,color="c",linewidth=3,label='Population of Vespa mandarinia')
plt.xlabel("t") #xlabel、ylabel：分别设置X、Y轴的标题文字。
plt.ylabel("N(t)")
plt.title("Population growth curve of Vespa mandarinia") # title：设置子图的标题。
plt.legend(loc='upper left')  #设置图例中显示的内容
plt.ylim(0,12000)# xlim、ylim：分别设置X、Y轴的显示范围。
plt.fill_between(t,N,0,color='blue',alpha=0.25)
plt.savefig('C:/Users/LENOVO/Desktop/S curve.eps',dpi=120,bbox_inches='tight')
plt.show()
```

```python
import time
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

t = np.linspace(0, 10, 1000)
N0=100
r=1.2
K=10000
N=K/(1+(K-N0)/N0*np.exp(-r*t))
r1=1.5
N1=K/(1+(K-N0)/N0*np.exp(-r1*t))
plt.figure(figsize=(6,4))
plt.plot(t,N,color="c",linewidth=3,label='Population of Vespa mandarinia')
plt.plot(t,N1,color="yellow",linewidth=3,label='Increase Growth Rate r')
plt.xlabel("t") #xlabel、ylabel：分别设置X、Y轴的标题文字。
plt.ylabel("N(t)")
plt.title("Population growth curve of Vespa mandarinia") # title：设置子图的标题。
plt.legend(loc='upper left')  #设置图例中显示的内容
plt.ylim(0,12000)# xlim、ylim：分别设置X、Y轴的显示范围。
plt.fill_between(t,N,N1,color='red',alpha=0.25)
plt.annotate("Vespa mandarinia Increment",
             xy=(3.8,6000), xytext=(6,4000), 
             weight="bold", color="black",
             arrowprops=dict(arrowstyle="->",
                             connectionstyle="arc3",
                             color="r"))
plt.savefig('C:/Users/LENOVO/Desktop/R+.eps',dpi=120,bbox_inches='tight')
plt.show()
```

```python
import time
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

t = np.linspace(0, 10, 1000)
N0=100
r=1.2
K=10000
N=K/(1+(K-N0)/N0*np.exp(-r*t))
r1=1
N1=K/(1+(K-N0)/N0*np.exp(-r1*t))
plt.figure(figsize=(6,4))
plt.plot(t,N,color="c",linewidth=3,label='Population of Vespa mandarinia')
plt.plot(t,N1,color="blue",linewidth=3,label='Decrease Growth Rate r')
plt.xlabel("t") #xlabel、ylabel：分别设置X、Y轴的标题文字。
plt.ylabel("N(t)")
plt.title("Population growth curve of Vespa mandarinia") # title：设置子图的标题。
plt.legend(loc='upper left')  #设置图例中显示的内容
plt.ylim(0,12000)# xlim、ylim：分别设置X、Y轴的显示范围。
plt.fill_between(t,N,N1,color='red',alpha=0.25)
plt.annotate("Vespa mandarinia Decrement",
             xy=(4.2,6000), xytext=(6,4000), 
             weight="bold", color="black",
             arrowprops=dict(arrowstyle="->",
                             connectionstyle="arc3",
                             color="r"))
plt.savefig('C:/Users/LENOVO/Desktop/R-.eps',dpi=120,bbox_inches='tight')
plt.show()
```
