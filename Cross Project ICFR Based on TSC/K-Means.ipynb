```python
#导入
import pandas as pd
import numpy as np
from sklearn import preprocessing
import os
from sklearn.cluster import KMeans
import math

#遍历文件
dfs=[] 
dirpath="D:\Desktop\ICFR\\" #设置读取路径，最后一个必须是\\
for root_dir,sub_dir,files in os.walk(dirpath): #遍历文件
    for file in files:
        df=pd.read_excel(dirpath+file,encoding='ISO-8859-1', sep='\t',header=0)
        df.eval('flag=flag-1',inplace=True) #设置flag为0，1
        dfs.append(df)


      
```

```python
#K-means
from sklearn import preprocessing

def Kmeans(data):
    #提取特征部分
    x= data.drop('flag', 1)
    
    #z-score标准化数据
    temp=[]
    for index,row in x.iteritems():
        temp.append(preprocessing.scale(row))
    for jj in range(0,89):
        x.iloc[:,jj]=temp[jj]
    
    
    estimator=KMeans(n_clusters=2)
    #构造聚类模型
    estimator.fit(x)
    #获取聚类标签
    y=estimator.labels_
    #合并为新数据
    data['pred']=y
    #抽取两个簇
    cluster0=data[(data['pred']==0)]
    cluster1=data[(data['pred']==1)]
    #计算均值
    sum0=0
    mean0=0
    for index, row in cluster0.iterrows():#遍历每一行
        sum0=sum0+row
    for i in range(0,89):
        mean0=mean0+sum0[i]
    sum1=0
    mean1=0
    for index, row in cluster1.iterrows():#遍历每一行
        sum1=sum1+row
        
    for i in range(0,89):
       
        mean1=mean1+sum1[i]
    
    if mean0<mean1:
        data['pred'==1]=2
        data['pred'==0]=1
        data['pred'==2]=0

```

```python
#计算机器学习性能

import math

def Performance_measure(data):
    #提取标签
    flag=data['flag']
    pred=data['pred']


    def calculate_TP(y, y_pred):
        tp = 0
        for i, j in zip(y, y_pred):
            if i == j == 1:
                tp += 1
        return tp
    def calculate_TN(y, y_pred):
        tn = 0
        for i, j in zip(y, y_pred):
            if i == j == 0:
                tn += 1
        return tn
    def calculate_FP(y, y_pred):
        fp = 0
        for i, j in zip(y, y_pred):
            if i == 0 and j == 1:
                fp += 1
        return fp
    def calculate_FN(y, y_pred):
        fn = 0
        for i, j in zip(y, y_pred):
            if i == 1 and j == 0:
                fn += 1
        return fn

    def calculate_TP2(y, y_pred):
        tp = 0
        for i, j in zip(y, y_pred):
            if i == j == 0:
                tp += 1
        return tp
    def calculate_TN2(y, y_pred):
        tn = 0
        for i, j in zip(y, y_pred):
            if i == j == 1:
                tn += 1
        return tn
    def calculate_FP2(y, y_pred):
        fp = 0
        for i, j in zip(y, y_pred):
            if i == 1 and j == 0:
                fp += 1
        return fp
    def calculate_FN2(y, y_pred):
        fn = 0
        for i, j in zip(y, y_pred):
            if i == 0 and j == 1:
                fn += 1
        return fn

    def calculate_MCC(TP,FP,FN,TN):
        numerator = (TP * TN) - (FP * FN) #马修斯相关系数公式分子部分
        denominator = math.sqrt((TP + FP) * (TP + FN) * (TN + FP) * (TN + FN)) #马修斯相关系数公式分母部分
        result = numerator/denominator
        return result

    TN=calculate_TN(flag,pred)
    TP=calculate_TP(flag,pred)
    FN=calculate_FN(flag,pred)
    FP=calculate_FP(flag,pred)

    TN2=calculate_TN2(flag,pred)
    TP2=calculate_TP2(flag,pred)
    FN2=calculate_FN2(flag,pred)
    FP2=calculate_FP2(flag,pred)

    #分类性能指标
    Accuracy = (TN + TP) / (TN + FP + FN + TP)
    Recall = TP / (TP + FN)
    Recall2 = TP2 / (TP2 + FN2)
    Pd = Recall
    Precision = TP / (TP + FP)
    Precision2 = TP2 / (TP2 + FP2)
    False_Positive = FP / (FP + TN)
    Pf = False_Positive
    F_measure = 2 * Precision * Recall / (Precision + Recall)
    F_measure2 = 2 * Precision2 * Recall2 / (Precision2 + Recall2)
    G_measure = 2 * Recall * (1 - False_Positive) / (Recall + (1 - False_Positive))
    g_mean = math.sqrt((TN / (TN + FP)) * (TP / (TP + FN)))
    Bal = 1- math.sqrt(math.pow(0-Pf,2)+math.pow(1-Pd,2))/math.sqrt(2)
    MCC=calculate_MCC(TP,FP,FN,TN)

    Performance=[Precision, Recall, Pf, F_measure, G_measure, g_mean, Bal, MCC,F_measure2]
    return Performance
```

```python
#统计7个数据集
result =pd.DataFrame()
   
for ii in range(0,7):
    data=dfs[ii]
    Kmeans(data)
    temp=pd.DataFrame()
    temp[ii]=Performance_measure(data)
    result=result.append(temp.T,ignore_index=True)
    
result.columns=["Precision","Recall","Pf","F_measure","G_measure","g_mean","Bal","MCC","F_measure2"]
result
#result.to_excel('D:\Desktop\K-means对比实验结果.xlsx')
```
