```python
import numpy as np
import pandas as pd
#导入数据
tables=[]
f=pd.read_excel('D:/Desktop/temp.xlsx','F')
G=pd.read_excel('D:/Desktop/temp.xlsx','G')
g=pd.read_excel('D:/Desktop/temp.xlsx','g-mean')
b=pd.read_excel('D:/Desktop/temp.xlsx','b')
m=pd.read_excel('D:/Desktop/temp.xlsx','m')
tables.append(f)
tables.append(G)
tables.append(g)
tables.append(b)
tables.append(m)
methods=['NN-Filter','Peter-Filter','Yu-Filter','HISNN','TCA+','TNB','CODEP','ASCI','TSC']
indicators=['F-measure','G-measure','g-mean','Balance','MCC']
```

```python
#箱线图
import seaborn as sns
import matplotlib.pyplot as plt
def drawBox(aaa):
    plt.figure(figsize=(10, 8))
    sns.set_theme(style='white',font='Times New Roman',font_scale=2.5)
    #设置调色
    ppp=sns.color_palette("Blues_r")
    ax=sns.boxplot(order=methods,data=tables[aaa],linewidth=2,palette=ppp)
    #单独设置颜色
    ax.artists[-1].set_facecolor('firebrick')
    ax.tick_params(axis='x',labelsize=10,colors='black') 
    ax.set_xticklabels(ax.get_xticklabels(), rotation=20,fontsize=20,fontweight='bold')
    
    plt.savefig('D:/Desktop/figures/box/'+indicators[aaa]+'.pdf', dpi=1000)

for aaa in range(0,5):
    drawBox(aaa)
```
