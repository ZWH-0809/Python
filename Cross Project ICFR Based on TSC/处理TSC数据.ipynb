```python
#导入
import pandas as pd
import numpy as np
import math
import os
from pandas import Series,DataFrame

#遍历文件

List=pd.DataFrame()

dirpath="D:\Desktop\实验结果\\" #设置读取路径，最后一个必须是\\
for root_dir,sub_dir,files in os.walk(dirpath): #遍历文件
    for file in files:
        df=pd.read_excel(dirpath+file)
        
        df.loc[-1] = df.columns.tolist()
        df.index = df.index + 1
        df.sort_index(inplace=True)
        df.columns = ["A","B","C","D","E","F","G","H","I","J","K","L","M","N","O","P","Q"]
        
        
        F1=df['D']
        
        List=pd.concat([List, F1], axis=1)
        List.rename(columns={'D':file},inplace=True)
       
List.insert(len(List.columns),'max',List.max(axis=1))             
List.insert(len(List.columns),'maxindex',List.idxmax(1))
#print(List)
print(List['maxindex'])
```

```python
names=List['maxindex']
Table=pd.DataFrame()

i=0
result =pd.DataFrame(columns=("A","B","C","D","E","F","G","H","I","J","K","L","M","N","O","P","Q"))
for item in names:
    datas=pd.read_excel(dirpath+item)
    
    datas.loc[-1] = datas.columns.tolist()
    datas.index = datas.index+1
    datas.sort_index(inplace=True)
    datas.columns = ["A","B","C","D","E","F","G","H","I","J","K","L","M","N","O","P","Q"]
    
    data=datas.iloc[i,:]
    result=result.append(data,ignore_index=True)

   
    
    i=i+1
    
#输出结果
result.columns = ["precision","recall","pf","F-measure","G-measure","g-mean","balance","MCC","accuracy2","precision2","recall2","F-measure2","G-measure2","g-mean2","bal2","MCC2","AUC"]
label=[]
for str in names:
    label.append(str[8:13])
result['λ_k ']=label
result    
result.to_excel('D:\Desktop\TSC实验结果汇总.xlsx')
    
```
