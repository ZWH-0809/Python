```python
import numpy as np
import pandas as pd

tables=[]
f=pd.read_excel('D:/Desktop/TSC-unsupervised.xlsx','F-measure')
G=pd.read_excel('D:/Desktop/TSC-unsupervised.xlsx','G-measure')
g=pd.read_excel('D:/Desktop/TSC-unsupervised.xlsx','g-mean')
b=pd.read_excel('D:/Desktop/TSC-unsupervised.xlsx','balance')
m=pd.read_excel('D:/Desktop/TSC-unsupervised.xlsx','mcc')

def prep(data):
    #删除冗余列
    data=data.drop(columns=['Source','sy'])
    #排除nan
    data.dropna(axis=0,how='any',inplace=True)
    return data
f=prep(f).reset_index(drop=True)
G=prep(G).reset_index(drop=True)
g=prep(g).reset_index(drop=True)
b=prep(b).reset_index(drop=True)
m=prep(m).reset_index(drop=True)
tables.append(f)
tables.append(G)
tables.append(g)
tables.append(b)
tables.append(m)
projects=['Codec','Collections','IO','Jsoup','JSqlParser','Mango','Ormlite-Core']
f
```

```python
#重组数据
results=[]

for i in range(0,7):
    temp=pd.DataFrame(columns=f.columns)
    for j in range(0,5):
        temp=temp.append(tables[j].iloc[i,:],ignore_index=True)
    results.append(temp)
    
results[3]
```
